## Creating My Personalized Sample

## Import pandas 

```python
import pandas as pd

# Load master dataset
df_wdi_full = pd.read_csv("wdi_dataset.csv")

# Define your student ID
student_id = "nf1028810"  # Replace with your own ID
SAMPLE_FRACTION = 0.15
SEED_DIGITS = 3 

# Derive random seed
seed = int(str(student_id)[-SEED_DIGITS:])

# Create your sample
df_sample = df_wdi_full.sample(frac=SAMPLE_FRACTION, random_state=seed).reset_index(drop=True)

# Save the sample
filename = f"student_sample_{student_id}.csv"
df_sample.to_csv(filename, index=False)

print(f"Your sample ({len(df_sample)} rows) has been saved as {filename}")
print(f"Seed used: {seed}")
```

**Interpretation:**

The code defines a standardized process for creating a personalized, reproducible sample for analysis. It uses the student ID to derive a random seed 810 from nf1028810 and then samples 15\% of the master dataset (df_wdi_full). This ensures that the generated sample is unique to the student but can be recreated consistently, providing a fixed dataset for all subsequent tasks.

## TASK 6 – Creating New Columns

```python
# Convert population to millions
population = df_sampled_data["Population"] / 1000000

# Compute median and 75th percentile GDP
median_gdp = df_sampled_data["GDPperCapita"].median()
quantile_gdp = df_sampled_data["GDPperCapita"].quantile(0.75)

# Create GDP category
def classify_gdp(gdp):
    if gdp < median_gdp:
        return "Low"
    elif gdp < quantile_gdp:
        return "Medium"
    else:
        return "High"

df_sampled_data["GDPCategory"] = df_sampled_data["GDPperCapita"].apply(classify_gdp)

# Count how many rows fall into each category
category_count = df_sampled_data["GDPCategory"].value_counts()

print("\nCounts per Category:")
print(category_count)
```

**Interpretation:**

New columns, PopulationMillions and GDPCategory, have been successfully added to the DataFrame. The GDPCategory column buckets countries based on whether their GDP per capita is below the global median (Low), above the 75th percentile (High), or in between (Medium), providing a discrete variable for economic comparisons.

## TASK 9 – NumPy Array Operations

Import Numpy

```python
import numpy as np

#9.1 	Compute min, max, mean, and range for LifeExpectancy.

# Convert column to NumPy array
life = df_sampled_data["LifeExpectancy"].to_numpy()

# Compute statistics
min_life = np.min(life)
max_life = np.max(life)
mean_life = np.mean(life)
range_life = max_life - min_life

print("Min Life Expectancy:", min_life)
print("Max Life Expectancy:", max_life)
print("Mean Life Expectancy:", mean_life)
print("Range:", range_life)
```

**Interpretation:**

The average life expectancy across all observations is approximately 68.79 years, with a wide range of 53.59 years between the minimum (31.51 years) and the maximum (85.09 years). This vast range indicates significant global disparities in health outcomes and development levels within the sample.

```python
#9.2 	Compute 25th, 50th, and 75th percentiles.

# Compute percentiles
num_25 = np.percentile(life, 25)
num_50 = np.percentile(life, 50)   
num_75 = np.percentile(life, 75)

print("25th Percentile:", num_25)
print("50th Percentile:", num_50)
print("75th Percentile:", num_75)
```

**Interpretation:**

The median life expectancy (50th percentile) is 70.695 years, meaning half of all country-year observations have a life expectancy below this value. The interquartile range (IQR), between 63.268 years (25th percentile) and 75.752 years (75th percentile), captures the middle 50\% of the data.

```python
#9.3 	Compute correlation between GDPperCapita and LifeExpectancy.

df_clean = df_sampled_data.dropna(subset=['GDPperCapita', 'LifeExpectancy'])

gdp = df_clean['GDPperCapita'].values
life = df_clean['LifeExpectancy'].values

correlation = np.corrcoef(gdp, life)[0, 1]
print("Correlation between GDPperCapita and LifeExpectancy:", correlation)
```

**Interpretation:**

The Pearson correlation coefficient of 0.5744 indicates a moderately strong positive linear relationship between a country's GDP per capita and its life expectancy. This suggests that as a country's wealth increases, there is a clear tendency for the average life expectancy of its population to also increase, reflecting the benefits of economic development on health outcomes.
